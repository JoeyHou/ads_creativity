```python
# def get_c(c):
#     if c == 0:
#         return math.pi 
#     if c == 1:
#         return 2
#     if c % 2 == 0:
#         num = 1
#         for i in range(1, c, 2):
#             num = num * i
#         dom = 1
#         for i in range(2, c+1, 2):
#             dom = dom * i
#         return math.pi * (num) / (dom)
#     if c % 2 == 1:
#         dom = 1
#         for i in range(1, c+1, 2):
#             dom = dom * i
#         num = 1
#         for i in range(2, c, 2):
#             num = num * i
#         return 2 * (num) / (dom)

def get_k(m):
    if m == -1:
        return 1
    if m == 0:
        return math.pi 
    if m == 1:
        return 2 * math.pi 
    if m % 2 == 0:
        k = math.pi * (math.pi * 2) ** (m / 2)
        for i in range(2, m+1, 2):
            k = k / i
    if m % 2 == 1:
        k = (math.pi * 2) ** ((m + 1) / 2)
        for i in range(1, m+1, 2):
            k = k / i 
    return k

def get_target_formula(n, k):
    p = math.e ** (-n * 0.2) * (12 / math.pi)
    summation = 0
    for i in range(k-1):
        summation += (get_k(i-1) * (math.e * math.sqrt(n) / (2 * math.pi))**i )
    return p * summation
```

```python
x = range(80, 110)
y = [get_target_formula(i, 8) for i in x]
sns.scatterplot(x=x, y=y)
```
